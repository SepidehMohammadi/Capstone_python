```python
import pandas as pd
import numpy as np
import os
import seaborn as sns

path = os.path.join('Resources', 'online_shoppers_intention.csv')

df = pd.read_csv(path)
```

```python
oneclick_Adminstrative_users = df[df['Administrative'] == 1]
```

```python
oneclick_Adminstrative_users_mean_Adminstrative_Duration = oneclick_Adminstrative_users['Administrative_Duration'].mean(axis = 0, skipna = True)
```

```python
oneclick_Adminstrative_users_mean_Adminstrative_Duration
```

```python
df['Administrative_Duration'].loc[ (df['Administrative_Duration'] == 0) & (df['Administrative'] == 1)] = oneclick_Adminstrative_users_mean_Adminstrative_Duration
```
